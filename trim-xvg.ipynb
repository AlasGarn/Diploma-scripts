```python
import os 
%matplotlib inline
import matplotlib.pyplot as plot
from scipy.signal import argrelextrema
import numpy as np
```

```python
os.chdir("/home/domain/data/kirill/AB/XOP/md/new/restart/")
```

```python
files = []
for a in os.walk('.'):
    fz = a[2]
    files.append([a[0]+"/"+x for x in fz if x.endswith("frames.xvg")])
files = [k for z in [f for f in files if len(f) > 0] for k in z]
print files
#files = [files[0]] # for test
```

```python
f = open(files[0], 'r')
lns = f.readlines()
f.close()
```

```python
xs,ys,zs = [],[],[]
for ln in lns:
    xs.append(float(ln.split()[0]))
    ys.append(float(ln.split()[1]))
    zs.append(float(ln.split()[2]))
    
xs,ys,zs = np.array(xs),np.array(ys),np.array(zs)
plot.plot(xs, ys)
```

```python
extr = argrelextrema(ys, np.less)
print len(extr[0])
```

```python
plot.plot(xs[extr[0]],ys[extr[0]])
```
