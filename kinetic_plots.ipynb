```python
import numpy as np
%matplotlib inline
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
def get_ready(a):
    return np.array([1/float(t) for t in a.split()])[1::2]
```

```python
t10 = "96	0.000660313 110	0.000732301 146	0.000662917 212	0.000763808 460	0.001021765"
t15 = "96	0.000327468 110	0.000425754 146	0.000521876 212	0.000697487 460	0.000658208"
t20 = "96	0.000203136 110	0.00022167 146	0.000246446 212	0.000376758 460	0.000481017"
t26 = "96	0.000352903 110	0.000450607 146	0.000604258 212	0.000726525 460 0.001080851"
t30 = "96	0.000588741 110	0.00065914 146	0.000765099 212	0.001053718 460	0.001178435"
t10 = get_ready(t10)
t15 = get_ready(t15)
t20 = get_ready(t20)
t26 = get_ready(t26)
t30 = get_ready(t30)
x = np.array([1/float(x) for x in [96,110,146,212,460]])
```

```python
fig, ax = plt.subplots(figsize=(6,5))
sns.set_style("white")
sns.despine()
sns.regplot(x,t10, ci=None, marker = "v", line_kws={"lw": 1, "color": "black"},scatter_kws={"s": 80, "color": "black"}, label='T = 283 K')
sns.regplot(x,t15, ci=None, marker = "*", line_kws={"lw": 1, "color": "black"},scatter_kws={"s": 80, "color": "black"}, label='T = 288 K')
sns.regplot(x,t20, ci=None, marker = "o", line_kws={"lw": 1, "color": "black"},scatter_kws={"s": 80, "color": "black"}, label='T = 293 K')
sns.regplot(x,t26, ci=None, marker = "s", line_kws={"lw": 1, "color": "black"},scatter_kws={"s": 80, "color": "black"}, label='T = 299 K')
sns.regplot(x,t30, ci=None, marker = "x", line_kws={"lw": 1, "color": "black"},scatter_kws={"s": 80, "color": "black"}, label='T = 303 K')
ax.legend(loc="upper left")
plt.ylabel(r"$1/k_{obs}$",fontsize = "14")
plt.xlabel(r"$1/C_{XOP},\ \mu M^{-1} $",fontsize = "14")
```

```python
s = "0.003526839	-6.874960066 0.003472222	-6.729297887 0.003337784	-5.84729618 0.003299241	-5.938823297"
sex = np.array([float(t) for t in s.split()])[1::2]
x = np.array([float(t) for t in s.split()])[0::2]
```

```python
sns.regplot(x,sex, ci=None, marker = "o", line_kws={"lw": 1, "color": "black"}, scatter_kws={"s": 80, "color": "black"}, label='T = 293 K')
plt.ylabel(r"$ln(k_{2})$",fontsize = "14")
plt.xlabel(r"$1/T,\ K^{-1} $",fontsize = "14")
```

```python
p, V = np.polyfit(x, sex, 1, cov=True)

print "x_1: {} +/- {}".format(p[0], np.sqrt(V[0][0]))

print "x_2: {} +/- {}".format(p[1], np.sqrt(V[1][1]))
```
